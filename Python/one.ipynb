```python
# Sets - Unordered collection of data. As it is unordered there are no indexes.
set1 = set()
set1 = {'A','B','C','D','E','E','H'}
set2 = {'A','B','D','E','F','G','H','I','J'}
set3 = {'H','I','J','K','L','M','N','O'}
set1.add('Z')
print(set1.difference(set2))
print(set2.difference(set1))

#remove random value from set
print(set1)
set1.pop()
print(set1)

set1.pop()
print(set1)

# concatination of two sets without including duplicate values
print(set1.union(set2))

#update the set1 by dummy input set
print(set1.update({'O','P','Q'}))

#union of two sets containing elemets present in either of two sets but not in both sets 
print(set1.symmetric_difference(set2))

#common between both sets
print(set1.intersection(set2))

```
