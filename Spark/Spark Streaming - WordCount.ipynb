```python
from pyspark import SparkConf,SparkContext
from pyspark.streaming import StreamingContext
```

### 1.Create SparkStreaming Context

```python
#Intitialize new spark session
config = SparkConf().setAppName("SparkSession")
sc = SparkContext.getOrCreate(conf= config)
```

```python
#Initialize Streaming context with Batch Interval,Here 5 indicates 5 Seconds
ssc = StreamingContext(sc,5)
```

```python
# Create a Dstream that will connect to URL with netcat and port number
lines = ssc.socketTextStream("localhost",9988)
words = lines.flatMap(lambda line:line.split(" "))
pairs = words.map(lambda word:(word,1))
wordcounts = pairs.reduceByKey(lambda x,y:x+y)
wordcounts.pprint()

```

```python
ssc.start()
ssc.awaitTermination()
```

```python
ssc.stop()
```

```python
sc.stop()
```
