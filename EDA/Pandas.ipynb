#### 7.Handling Missing Values
    *Data Manipulation/Data Wrangling

```python
import numpy as np
```

```python
# replace '?' by 'nan' and update dataframe
raw_data.replace('?',np.nan,inplace=True)
```

```python
raw_data.isnull().sum()   #returns total missing values from each column name
```

```python
#Display missing value %
raw_data.isnull().sum()/raw_data.shape[0]*100
```

```python
#axis=1, Drop columns using column names
raw_data.drop(['symboling','normalized-losses'], axis=1, inplace=True)
```

```python
raw_data['price'].unique()
```

```python
raw_data['price']=raw_data['price'].astype('float64')
```

```python
vehicle_df = raw_data.copy()
```

```python
vehicle_df['price'].median()
```

```python
#fill missing value by median of 'price'
vehicle_df['price']=vehicle_df['price'].fillna(vehicle_df['price'].median())
vehicle_df
```

```python
# drop missing rows from all columns
vehicle_df.dropna(inplace=True)
```

```python
vehicle_df.isnull().sum()
```

```python
vehicle_df.isnull()
```

```python
#mean - continous
#median  - discrete and continuos
#mode -  categorical
```

#### 8.Drop Duplicate Rows
    * Drop duplicate values

```python
vehicle_df.duplicated()
```

```python
vehicle_df[vehicle_df.duplicated()]
```

```python
dict(vehicle_df.loc[10])
```

```python
#select random samples from dataframe using frac=0.05(5%) or n = num of samples
sample_df = vehicle_df.sample(frac=0.05)
```

```python
#inserting data
vehicle_df1= pd.concat([vehicle_df,sample_df],ignore_index=True)
```

```python
vehicle_df1[vehicle_df1.duplicated()]
```

```python
vehicle_df1.drop_duplicates(inplace=True)    #drop all duplicate rows from dataframe
```

#### 9.Group By
