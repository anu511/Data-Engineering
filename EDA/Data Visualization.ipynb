#### Univariate Analysis
* Univariate Analysis is the simplest form of analysis
* Only one variable is used in visualization.This doesn't deal with relationship plots.
* It takes data,summarize the data,patterns analysis,trend analysis in data.Univariate analysis can be implemented on "numerical data" or
  "categorical data"
  

```python
df_numerical = housing_data.select_dtypes(include=['number'])
df_numerical.info()
```

```python
df_categorical = housing_data.select_dtypes(include=['object'])
df_categorical.info()
df_categorical
```

#### 1.Line-Plot
* Univariate Data - Trends Analysis
* Bivariate Data - Relationship Plot

```python
#Plot using DataFrame and plot type = 'line'
df_numerical['SalePrice'].plot(kind='line')
```

```python
df_numerical['SalePrice'].sort_values(ascending=True).unique()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(5,4))
plt.plot(df_numerical['SalePrice'].sort_values(ascending=True).unique(),color='orangered')
plt.xlabel('Index')
plt.ylabel('Sale Price')
plt.title('Trends of Sale Price',size=10)   #size= set the font size for Title and Labels
plt.show()
```

#### 2.Bar Plot
* Univariate or Bivariate Plots
* Bar charts are used with Categorical Dataset
* Bar Plots are used to display counts of unique value for categorical type,Height of Bar indicate count for each categorical value

```python
df_categorical['MoSold'].value_counts()
```

```python
df_categorical['MoSold'].value_counts().plot(kind = 'bar',figsize = (5,4))
```

```python
plt.style.use('ggplot')
plt.bar(x=df_categorical['MoSold'].value_counts().index,  
        height=df_categorical['MoSold'].value_counts().values,color='green')
plt.xlabel('Months')
plt.ylabel('MoSold')
plt.title('Trends of Sale Price',size=10)   #size= set the font size for Title and Labels
plt.show()
```

#### 3.Histogram
* Univariate Analysis - Apply on continous data
* Frequency of Data in Range.
* Study of Spread/Data Distribution,

```python
#bins=20,Indicates 20 equal distribution of Numerical Dala
sns.set_style('whitegrid')
plt.figure(figsize=(10,8))
plt.hist(df_numerical['SalePrice'],bins=20,color='green')
plt.title("Histogram Sales Price")
```

#### 4.Subplots
* Here one plot will be splitted into multiple plots.Each index position will be occupied by one Plot.

```python
data=housing_data.query("LandSlope == 'Gtl' ")
data['SalePrice']
```

```python

plt.subplot(2,2,1)
plt.hist(housing_data[housing_data['LandSlope']=="Gtl"]['SalePrice'],bins=15,color='b',alpha=0.5)

plt.subplot(2,2,2)
plt.hist(housing_data[housing_data['LandSlope']=="Mod"]['SalePrice'],bins=15,color='r',alpha=0.5)

plt.subplot(2,2,3)
plt.hist(housing_data[housing_data['LandSlope']=="Sev"]['SalePrice'],bins=15,color='g',alpha=0.5)

plt.show()
```

#### 5.Scatter Plot
* Indicates relationship plot or correlate between two numerical columns

```python
plt.figure(figsize = (5,4))
plt.scatter(x=df_numerical[df_numerical['LotArea']<50000]['LotArea'],
            y=df_numerical[df_numerical['LotArea']<50000]['SalePrice'],color='g')
plt.xlabel('LotArea')
plt.ylabel('SalePrice')
plt.title('Scatter Plot - Relationship B/W LotArea and SalePrice')
```

```python
sns.scatterplot(x=df_numerical[df_numerical['LotArea']<50000]['LotArea'],
                y=df_numerical[df_numerical['LotArea']<50000]['SalePrice'],
                hue=df_categorical['GarageType'],s=50)
plt.show()
```

```python
sns.scatterplot(x=df_numerical['GarageArea'],
                y=df_numerical['SalePrice'],
                hue=df_categorical['GarageType'],s=50,marker = '*',alpha=0.7)  #s = size of bubble,marker = s,^,>,<,*,.,+,d(shape)
plt.show()
```

#### 6.Pie-Chart
* Frequency Distribution Plot

```python
plt.figure(figsize=(8,8))
plt.pie(df_categorical['LotConfig'].value_counts(),autopct = '%0.2f%%',labels =['Inside','Corner','CulDSac','FR2','FR3'])
plt.legend(loc=1)
plt.show()
```
