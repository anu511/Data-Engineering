```python
import pandas as pd

pd.Series(list)
```

```python
first_series = pd.Series(list('abcdefg'))
print(first_series)
```

```python
import numpy as np
city_series = pd.Series(np.array(['Luxemburg','Ireland','Singapore','Norway','Qatar',
                                   'United Arab Emirates','Denmark','Sweden']))
print(city_series)
```

```python
#By default index position will be auto-incremented with start=0 and end=8
#to change index position explicitly
city_series = pd.Series(np.array(['Luxemburg','Ireland','Singapore','Norway','Qatar',
                                   'United Arab Emirates','United States','Denmark','Sweden']),
                                   index=['a','b','c','d','e','f','g','h','i'])
print(city_series)
```

```python
# Collection of city_series
country_per_capita = {'Country': pd.Series(np.array(['Luxemburg','Ireland','Singapore','Norway','Qatar',
                                   'United States','Denmark','Sweden'])),
 'GDP_Per_Capita':pd.Series(np.array([142212,126905,127565,114899,114648,76399,74005,64578]))
}
#Creating DataFrame using Collections of Series
GDP_DF= pd.DataFrame(country_per_capita)

#Display first 5 rows
print(GDP_DF.head())
```
