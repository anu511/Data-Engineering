```python
class Expense:
    def __init__(self,expense_id,date,category,description,amount):
        self.expense_id=expense_id
        self.date=date
        self.category=category
        self.description=description
        self.amount=amount
    
    def __str__(self):
        return f'Expense ID: {self.expense_id} Date: {self.date} Category: {self.category} Description: {self.description} Amount: {self.amount}'

expenses=[]


def add_expense(expense):
    expenses.append(expense)
    print("Expense added successfully")

def update_expense(expense_id,new_expense):
    for exp in expenses:
        if exp.id==expense_id:
            exp.date=new_expense.date
            exp.category=new_expense.category
            exp.description=new_expense.description
            exp.amount=new_expense.amount
            print("Expense updated sucessfully")
            return
    print("Expense not found")

def delete_expense(expense_id):
    for item in expenses:
        if item.expense_id==expense_id:
            del expenses[item]
            print("Expense deleted !")

def display_expense():
    print()
    for item in expenses:
        print(item)

users={'admin':'admin@123','anu':'ust@123','user':'user@123'}

def authenticate_user(username,password):
    if username in users and users[username]==password:
        print("Authentication Successfull\n")
        return True
    else:
        print("Authentication failed.Invalid username or password\n")
        return False

def categorize_expense():
    categories={}
    for exp in expenses:
        if exp.category in categories:
            categories[exp.category]+=int(exp.amount)
        else:
            categories[exp.category]=int(exp.amount)
    return categories

def summarize_expenses():
    total=0
    for exp in expenses:
        total+=int(exp.amount)
    return total

def generate_summary_report():
    catexp=categorize_expense()
    print("\nExpense Summary Report\n")
    for category,amount in catexp.items():
        print(f"Category :{category},Total Amount {amount}")
    tot=summarize_expenses()
    print("Total Expense: ",tot)
    print()

def cli():
    username=input("Enter Username: ")
    password=input("Enter Password: ")

    if(authenticate_user(username,password)):

        while True:
            print("----MENU----")
            print("1.Add a new expense")
            print("2.Update an Expense")
            print("3.Delete Expense")
            print("4.Display")
            print("5.Generate a summary report")
            print("6.Exit")

            choice=input("Enter Choice: ")
            
            if choice=='1':
                id=input("Enter expense id: ")
                date=input("Enter date: ")
                category=input("Enter category: ")
                description=input("Enter description: ")
                amount=input("Enter amount: ")
                expense=Expense(id,date,category,description,amount)
                add_expense(expense)

            elif choice=='2':
                expense_id=input("Enter expense id to update: ")
                date=input("Enter new date: ")
                category=input("Enter new category: ")
                description=input("Enter new description: ")
                amount=input("Enter new amount: ")
                new_expense=Expense(expense_id,date,category,description,amount)
                update_expense(expense_id,new_expense)

            elif choice=='3':
                expense_id=input("Enter expense id to delete: ")
                delete_expense(expense_id)

            elif choice=='4':
                display_expense()

            elif choice=='5':
                generate_summary_report()

            elif choice=='6':
                print("Exiting.....")
                break
            else:
                print("Invalid Choice")
            

cli()

```
