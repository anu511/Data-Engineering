```python
user_db={'user1':'1234','user2':'5678','user3':'91011'}
admin_db={'admin1':'987','admin2':'654','admin3':'123'}
```

```python
def user_login():
    username=input("Enter username: ")
    password=input("Enter password: ")
    if username in user_db and user_db[username]==password:
        print("\nLogined Successfully")
        return True
    else:
        print("\nInvalid Username or Password")
        return False
```

```python
def admin_login():
    username=input("Enter username: ")
    password=input("Enter password: ")
    if username in admin_db and admin_db[username]==password:
        print("\nLogined Successfully")
        return True
    else:
        print("\nInvalid Username or Password")
        return False
```

```python
def generate_session_id():
    return "12345"
```

```python
catalog=[{'id':'10','name':'Boots','category':'Footwear','price':2000},
         {'id':'20','name':'Coat','category':'Clothing','price':1000},
         {'id':'30','name':'Jacket','category':'Clothing','price':1500},
         {'id':'40','name':'Phone','category':'Electronics','price':50000}]
```

```python
def display_catalog():
    print("ID\tName\tCategory\tPrice")
    for item in catalog:
        print(f"{item['id']}\t {item['name']}\t{item['category']}\t{item['price']}")
        print()
```

```python
cart=[]

def display_cart():
    print()
    print("SID\tPID\tQuantity")
    for c in cart:
        print(f"{c['SID']} \t {c['PID']}\t {c['Quantity']}")
```

```python

def add_to_cart():
    sessionID=generate_session_id()
    productID=input("Enter product ID: ")
    quantity=int(input("Enter quantity: "))
    cart.append({'SID':sessionID,'PID':productID,'Quantity':quantity})
```

```python
def delete_from_cart():
    sessionID=generate_session_id()
    productID=input("Enter product ID: ")
    for item in cart:
        if item['PID'] == productID:
            cart.remove(item)
            print(f"Product with ID {productID} has been removed from the cart.")
            return
```

```python
def add_products():
    pid=input("Enter product ID: ")
    name=input("Enter Product Name: ") 
    category = input("Enter category: ")
    price = input("Enter price: ")
    catalog.append({'id':pid,'name':name,'category':category,'price':price})
```

```python
def update_products():
    pid=input("Enter product ID: ")
    name=input("Enter Product Name: ") 
    category = input("Enter category: ")
    price = input("Enter price: ")
    for product in catalog:
        if product['id']==pid:
            product['name']=name if name else product['name']            
            product['category']=category if category else product['category']  
            product['price']=price if price else product['price']  
            print(f'Product Updated')
```

```python
def delete_products():
    pid=input("Enter product ID to delete: ")
    for product in catalog:
        if product['id']==pid:
            catalog.remove(product)
            print("Product Deleted")
        
```

```python
def checkout():
    print("\nPayment Options")
    ch2=input("1.NetBanking \n2.Paypal\nEnter choice: ")
    print("You will be shortly redirected to the portal....")
    cart.clear()
    print("Your order is successfully placed") 
```

```python
while True:
    print("1.User")
    print("2.Admin")
    print("3.Exit")
    choice=input("Enter choice: ")
    if choice=='1':
        if(user_login()):
        
            while True:
                print()
                print("MENU")
                print("1.Display Cart")
                print("2.Add to Cart")
                print("3.Delete from Cart")
                print("4.Display Catalogue")
                print("5.Checkout")
                print("6.Log Out")

                ch=input("Enter your choice: ")

                if ch=='1':
                    display_cart()
                elif ch=='2':
                    add_to_cart()
                elif ch=='3':
                    delete_from_cart()
                elif ch=='4':
                    display_catalog()
                elif ch=='5':
                    checkout()
                elif ch=='6':
                    print("Logging out.....\n")
                    break
                else:
                    print("Invalid Choice ")
                    break

    elif choice=='2':
        if(admin_login()):

            while True:
                print("Admin Functions")
                print("1.Add Products")
                print("2.Update Products")
                print("3.Delete Products")
                print("4.Display Products")
                print("5.Exit")

                ch1=input("Enter choice")
                if ch1=='1':
                    add_products()
                elif ch1=='2':
                    update_products()
                elif ch1=='3':
                    delete_products()
                elif ch1=='4':
                    display_catalog()
                else:
                    print("Exiting.....\n")
                    break

    elif choice=='3':
        print("Exiting......")
        break
    else:
        print("Invalid Choice")
        break
```
